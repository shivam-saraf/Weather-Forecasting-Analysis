```python
#remove the commas from data frame
columns = weather_df.columns
for i in columns:
    weather_df[i] = weather_df[i].apply(lambda X:X.replace("'",""))  
```

```python
weather_df = weather_df[weather_df['Date'] != '2022-02-29']
```

```python
# change the data type of Date from object to datetime
weather_df["Date"] = pd.to_datetime(weather_df["Date"])
```

```python
# remove the irrelevent data 
weather_df = weather_df[weather_df[' diff_pressure'] != '0.11700000000000088;']
```

```python
# change the data type of columns from object to float
weather_df[weather_df.columns[weather_df.columns!='Date']] = weather_df[weather_df.columns[weather_df.columns!='Date']].astype(float)
```

```python
# change the data type of Month from float to int
weather_df[' Month']=weather_df[' Month'].astype(int)
```
